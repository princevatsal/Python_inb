```python
# K-fold Cross Validation
#Prepare X_test, X_train and y_test,y_train
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def build_classifier():
    classifier=Sequential()
    classifier.add(Dense(output_dim=6,init='uniform',activation='relu',input_dim=11))
    classifier.add(Dropout(rate=0.1))
    classifier.add(Dense(output_dim=6,init='uniform',activation='relu'))
    classifier.add(Dropout(rate=0.1))
    classifier.add(Dense(output_dim=1,init='uniform',activation='sigmoid',))
    classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
```

```python
classifier_sklearn=KerasClassifier(build_fn=build_classifier,batch_size=10,epochs=100)
```

```python
accuries=cross_val_score(estimator=classifier_sklearn,X=X_train,y=y_train,cv=10,n_jobs=-1)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
def build_classifier(optimizer):
    classifier=Sequential()
    classifier.add(Dense(output_dim=6,init='uniform',activation='relu',input_dim=11))
    classifier.add(Dropout(rate=0.1))
    classifier.add(Dense(output_dim=6,init='uniform',activation='relu'))
    classifier.add(Dropout(rate=0.1))
    classifier.add(Dense(output_dim=1,init='uniform',activation='sigmoid',))
    classifier.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
```

```python
classifier_sklearn=KerasClassifier(build_fn=build_classifier)
```

```python
parameters={
    'batch_size':[25,32],
    'epochs':[100,500],
    'optimizer':["adam","rmsprop"]
}
```

```python
grid_search=GridSearchCV(estimator=classifier_sklearn,
                         param_grid=parameters,
                         scoring='accuracy',
                         cv=10)
```

```python
grid_search=grid_search.fit(X_train,y_train)
```

```python
best_parameters=grid_search.best_params_
best_accuracy=grid_search.best_score_
```
